```python
import mysql.connector as connection
import pandas as pd
```

```python
sql = "SELECT * FROM classicmodels.orders;"
```

```python
# username & password should be same as that of creating a connection in docker container
# & while creating connection in MySQL UI
db = connection.connect(host="localhost", database="classicmodels", user="root", passwd="demopass01", use_pure=True)
```

```python
db
```

```python
df = pd.read_sql(sql, db)
```

```python
df
```

```python
df.head()
```

```python
# This will write/save EXCEL at same location as that of *.ipynb file
df.to_excel("mysql_date.xlsx", index=False)
```

```python
# We should close the connection which has been opened
db.close()
```
